```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import AdaBoostClassifier
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
data = pd.read_csv('spambase\spambase.data').as_matrix()
np.random.shuffle(data)
```

```python
X = data[:,:48]
Y = data[:,-1]
```

```python
X_train = X[:-100,:]
Y_train = Y[:-100]
X_test = X[-100:,:]
Y_test = Y[-100:]
```

```python
def model_fit(X_train,Y_train,X_test,Y_test,classifier='NiaveBayes'):
    if classifier=='Adaboost':
        model = AdaBoostClassifier()
        model.fit(X_train,Y_train)
        print("Classificaition Rate for AdaBoost:-",model.score(X_test,Y_test))
        return model
    else:
        model = MultinomialNB()
        model.fit(X_train,Y_train)
        print("Classificaition Rate for NB:-",model.score(X_test,Y_test))
        return model
```

```python
model_fit(X_train,Y_train,X_test,Y_test)
```

```python
model_fit(X_train,Y_train,X_test,Y_test,classifier='Adaboost')
```

# SMS

```python
data_kaggle_sms = pd.read_csv('spam-filter\spam.csv',encoding='ISO-8859-1')
data_kaggle_sms =data_kaggle_sms.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
data_kaggle_sms.columns =['label','data']
```

```python
data_kaggle_sms['b_label'] = data_kaggle_sms['label'].map({'ham': 0, 'spam': 1})
Y = data_kaggle_sms['b_label'].values
```

```python
count_vectorizer = CountVectorizer(decode_error='ignore')
X = count_vectorizer.fit_transform(data_kaggle_sms['data'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1)
```

```python
model = model_fit(X_train,Y_train,X_test,Y_test)
```

```python
# visualize the data
def visualize(df,label):
    words = ''
    for msg in df[df['label'] == label]['data']:
        msg = msg.lower()
        words += msg + ' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
visualize(data_kaggle_sms,'spam')
```

```python
visualize(data_kaggle_sms,'ham')
```

```python
data_kaggle_sms['predictions'] = model.predict(X)
```

```python
# things that should be spam
sneaky_spam = data_kaggle_sms[(data_kaggle_sms['predictions'] == 0) & (data_kaggle_sms['b_label'] == 1)]['data']
for msg in sneaky_spam:
    print(msg)
```

```python
# things that should not be spam
not_actually_spam = data_kaggle_sms[(data_kaggle_sms['predictions'] == 1) & (data_kaggle_sms['b_label'] == 0)]['data']
for msg in not_actually_spam:
    print(msg)
```

```python
cm = confusion_matrix(data_kaggle_sms['b_label'],data_kaggle_sms['predictions'])
```
